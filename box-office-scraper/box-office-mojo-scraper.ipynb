```python
# import libraries
import pandas as pd
```

```python
# get all tables
tables10 = pd.read_html("https://www.boxofficemojo.com/yearly/chart/?yr=2010&view=releasedate&view2=domestic&sort=gross&order=DESC&&p=.htm")
```

```python
movies10 = tables10[3]
```

```python
movies10 = movies10[5:]
movies10 = movies10[:10]
movies10 = movies10.iloc[:,0:9]
movies10
```

```python
movies10.to_csv(path_or_buf="movies10.csv")
```
