```python
from lxml import html
import requests

import pandas as pd
```

```python
# AMERICAN SNIPER
page = requests.get('https://www.boxofficemojo.com/movies/?page=daily&id=americansniper.htm')
tree = html.fromstring(page.content)
```

```python
american_sniper_daily = tree.xpath('//font[@color="#000080"]/text()')
```

```python
american_sniper_daily
```

```python
american_sniper_daily = pd.DataFrame(american_sniper_daily)
american_sniper_daily.to_csv(path_or_buf="american_sniper_daily.csv")
```
